```python
from sklearn.datasets import load_iris
# the iris_dataset return by `load_iris` is a Bunch object, similar to a dict with key/value pairs
iris_dataset = load_iris()
print("keys of iris dataset:", iris_dataset.keys())
```

```python
for (k, v) in iris_dataset.items():
    print(k, v)
```

```python
# data
# type() is a BIF
print('type of data', type(iris_dataset['data']))
# where did this `shape` coming from ?
print('shape of data', iris_dataset['data'].shape)

# target
print('shape of target', iris_dataset['target'].shape)
# The list slice access method
print('first 5 columns of target', iris_dataset['target'][0:5])
```

```python
from sklearn.model_selection import train_test_split
# this func takes 1st arg as dataset
# 2nd arg as result (target)
# and the random_state is to make sure that we have the same output whenever we run this function with the same input
# it's called fixed seed of pseudorandom number generator 
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state = 0)
print('training data shape', X_train.shape)
print('training target shape', y_train.shape)
```

```python
# create a DataFrame from data in X_train
# label columns with the names listed in feature_names
iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset['feature_names'])
# print('the data frame', iris_dataframe)
# create a scatter martix from the dataframe, color by y_train
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(10, 10),
                           marker='o', hist_kwds={'bins': 20}, s=60,
                           alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# we use only single neighbor this time
# this method returns an object that contains the algo to build model, and the algo to predicte new data points
knn = KNeighborsClassifier(n_neighbors=1)
# the fit method inside knn returns the knn object itself, with modifications in place
# feed it with our input training set, and corresponding output training set
knn.fit(X_train, y_train)
```

```python
# pretend we have found some new iris in the wild
# for now we have only one, but remember it should be a 2-dim array that matches our training dataset
new_X = np.array([[5, 2.9, 1, 0.2]])
predict = knn.predict(new_X)
print('prediction:', predict)
print('predicted target name:', iris_dataset['target_names'][predict])
```

```python
prediction = knn.predict(X_test)
print('predictions on test set:', prediction)

# here comes the magic np things....
score = np.mean(prediction == y_test)
print('test score:', score)

# or even better, the build-in method right in knn object
score_k = knn.score(X_test, y_test)
print('score directly from knn: ', score_k)
```
